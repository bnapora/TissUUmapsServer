## Load the TissUUmaps juptyer module

```python
import tissuumaps.jupyter as tj
```

## Display images and markers in TissUUmaps

```python
imageFiles = [
    "./data/breast/breastcancerHE.png",
    "./data/breast/breastcancerFluo.png"
]

csvFiles = [
    "./data/breast/breastcancer.csv"
]
```

#### Display images and csv files with some marker parameters:

```python
tj.loaddata(imageFiles, csvFiles, xSelector="global_X_pos", ySelector="global_Y_pos",
        keySelector="letters", nameSelector="gene_name",
        scaleSelector="global_Y_pos", colormap="interpolateOranges", colorSelector="global_X_pos",
        boundingBox=[0.2,0.1,0.6,0.6], scaleFactor=0.002, 
        height=400, tmapFilename="mainProject")
```

#### Display images only:

```python
tj.loaddata(imageFiles, height=400)
```
